# https://facebook.github.io/prophet/docs/diagnostics.html

```python
import pandas as pd
from fbprophet import Prophet
df = pd.read_csv('example_wp_log_peyton_manning.csv')


```

```python
df.head()
```

```python
m = Prophet()
forecast = m.fit(df).predict(m.make_future_dataframe(periods=30))
fig1 = m.plot(forecast)

```

```python
from fbprophet.diagnostics import cross_validation

# By default, the initial training period is set to three times the horizon, and cutoffs are made every half a horizon
df_cv = cross_validation(m, initial='730 days', period='180 days', horizon = '365 days')
df_cv
```

```python
df_cv.groupby("cutoff").first()
```

```python
df_cv.groupby("cutoff").last()
```

```python
# mean squared error (MSE), 
# root mean squared error (RMSE), 
# mean absolute error (MAE), 
# mean absolute percent error (MAPE)

from fbprophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p
```

```python
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv, metric='mape')
```

```python
fig = plot_cross_validation_metric(df_cv, metric='mse')
```
